# Installation

MolPy is a pure Python package. Install it with pip:

```bash
pip install molcrafts-molpy
```

That's it! You are ready to go.

## Optional Dependencies

For additional features, install optional groups:

| Group | Command | Features |
|-------|---------|----------|
| **All** | `pip install molcrafts-molpy[all]` | Everything below |
| **Dev** | `pip install molcrafts-molpy[dev]` | Testing, linting, formatting |
| **Docs** | `pip install molcrafts-molpy[doc]` | Building documentation |
| **RDKit** | `pip install molcrafts-molpy[rdkit]` | Chemical informatics support |

## Verify Installation

````python
import molpy as mp

print(f"✅ MolPy version: {mp.__version__}")
print(f"   Installation path: {mp.__file__}")
````
